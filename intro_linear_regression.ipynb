## using boston dataset of sklearn

```
from sklearn import datasets
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

## loading dataset

```
boston=datasets.load_boston()
boston
```

## creating dataframe

```
df=pd.DataFrame(boston.data,columns=boston.feature_names)
df['price']=boston.target
df
```

## visualization

```
plt.hist(df['price'])
```

## train and test data

## Simple_linear_regression

```
rooms=df.RM
cost=df.price
rooms=np.array(rooms).reshape(-1,1)
cost=np.array(cost).reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(rooms,cost,test_size=.2,random_state=5)#random_state -so that each time you dont get random result
x_train
```

## model

```
sim_lin_reg=LinearRegression()
sim_lin_reg.fit(x_train,y_train)
y_pred=sim_lin_reg.predict(y_test)
rmsc=np.sqrt(mean_squared_error(y_test,y_pred))
r2=round(sim_lin_reg.score(x_test,y_test),2)
print(r2)
```


## visualization
---



```
space=np.linspace(min(rooms),max(rooms)).reshape(-1,1)
plt.scatter(rooms,cost)
plt.plot(space,sim_lin_reg.predict(space))
plt.xlabel("rooms")
plt.ylabel("cost")
space.shape
```

## multiple_linear_regession

```
y=df['price']
x=df.drop('price',axis=1)
x
```

## train and test data

```
x1,x2,y1,y2=train_test_split(x,y,test_size=.2,random_state=5)
x1
```

## creating the model

```
reg_all=LinearRegression()
reg_all.fit(x1,y1)
predicted_values=reg_all.predict(x2)
```

## evaluate the model

```
rmse=np.sqrt(mean_squared_error(y2,predicted_values))
rmse
```

```
r2=round(reg_all.score(x2,y2),2)
r2
```

## visualize

```
plt.scatter(y2,predicted_values)
plt.xlabel("actual value")
plt.ylabel("predicted value")
plt.xticks(range(0,int(max(y2)),4))
plt.yticks(range(0,int(max(y2)),4))
```
